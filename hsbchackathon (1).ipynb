T-statistic (47.73): This number tells you how different the average amount is between fraudulent and non-fraudulent transactions, relative to how much variability there is within each group. A very large T-statistic (like 47.73) indicates a substantial difference in the average amount between the two groups.

P-value (0.000): This tells you the probability that you would see such a big difference in amount between the two groups if there was actually no real difference in the population. A P-value of 0.000 means this probability is extremely low (virtually zero).

```python
# Step 8: Build null hypothesis if needed for statistical understanding (optional).
print("\nNull Hypothesis")
# This step is typically for statistical analysis and hypothesis testing, not direct data transformation.
# Example:
if 'amount' in df.columns and 'fraud' in df.columns:
      fraud_amounts = df[df['fraud'] == 1]['amount']
      non_fraud_amounts = df[df['fraud'] == 0]['amount']
      # Check if both groups have enough data for t-test
      if len(fraud_amounts) > 1 and len(non_fraud_amounts) > 1:
          t_stat, p_value = stats.ttest_ind(fraud_amounts, non_fraud_amounts, equal_var=False)
          print(f"T-statistic (amount vs fraud): {t_stat:.2f}, P-value: {p_value:.3f}")
      else:
          print("Not enough data in fraud/non-fraud groups for statistical test on 'amount'.")
```

The strong T-statistic and extremely low P-value mean that amount is a very useful feature for distinguishing between fraudulent and non-fraudulent transactions. It suggests that the transaction amount is a powerful indicator for fraud detection in your dataset.

SMOTE (Synthetic Minority Oversampling Technique) generates synthetic samples for the minority class in imbalanced datasets by interpolating between existing samples and their nearest neighbors. It balances class distribution to improve model performance in detecting rare events.

```python
print(f"\nOriginal X shape: {X.shape}, y shape: {y.shape}")
print(f"Original class distribution: {y.value_counts()}")

if len(y.unique()) > 1 and y.value_counts().min() > 0:
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)

    print(f"\nResampled X shape: {X_resampled.shape}, y shape: {y_resampled.shape}")
    print(f"Resampled class distribution: {y_resampled.value_counts()}")
else:
    print("\nCannot apply SMOTE: Either only one class present or minority class has 0 samples.")
    X_resampled, y_resampled = X, y
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, precision_recall_curve, ConfusionMatrixDisplay

from hpelm import ELM
X = df_cleaned.drop('fraud', axis=1)
y = df_cleaned['fraud']

# SMOTE for class imbalance
if len(y.unique()) > 1 and y.value_counts().min() > 0:
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)
else:
    X_resampled, y_resampled = X, y

# Train-test split
if len(y_resampled.unique()) > 1 and y_resampled.value_counts().min() >= 2:
    X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, stratify=y_resampled, random_state=42)
else:
    X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Convert to NumPy arrays and explicitly ensure float type for ELM
X_train_np = X_train.to_numpy().astype(float)
y_train_np = y_train.to_numpy().reshape(-1, 1).astype(float) # Ensure target is float too
X_test_np = X_test.to_numpy().astype(float)


# --- Function to Train and Evaluate ELM with different parameters ---
def train_and_evaluate_elm(n_neurons, activation_function, X_train_data, y_train_data, X_test_data, y_test_true):
    print(f"\n--- ELM Configuration: {n_neurons} neurons, '{activation_function}' activation ---")

    elm = ELM(X_train_data.shape[1], 1)
    elm.add_neurons(n_neurons, activation_function)
    print(f"Training ELM with {n_neurons} neurons and '{activation_function}' activation...")
    elm.train(X_train_data, y_train_data)
    print("ELM training complete.")

    y_pred_proba_elm = elm.predict(X_test_data)
    y_pred_class_elm = (y_pred_proba_elm > 0.5).astype(int)

    # --- Metrics Evaluation ---
    print("\n--- ELM Evaluation Metrics ---")
    precision = precision_score(y_test_true, y_pred_class_elm)
    recall = recall_score(y_test_true, y_pred_class_elm)
    f1 = f1_score(y_test_true, y_pred_class_elm)
    roc_auc = roc_auc_score(y_test_true, y_pred_proba_elm)

    print(f"Precision: {precision:.4f}")
    print(f"Recall (Sensitivity): {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"AUC-ROC: {roc_auc:.4f}")

    # --- Visualizations ---
    print("\n--- ELM Model Visualizations ---")

    plt.figure(figsize=(7, 6))
    cm_display = ConfusionMatrixDisplay.from_predictions(y_test_true, y_pred_class_elm, cmap='Blues', values_format='d')
    cm_display.ax_.set_title(f'Confusion Matrix (ELM - {n_neurons} {activation_function})')
    plt.show()

    fpr, tpr, _ = roc_curve(y_test_true, y_pred_proba_elm)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic (ROC) Curve (ELM - {n_neurons} {activation_function})')
    plt.legend(loc="lower right")
    plt.show()

    precision_vals, recall_vals, _ = precision_recall_curve(y_test_true, y_pred_proba_elm)
    plt.figure(figsize=(8, 6))
    plt.plot(recall_vals, precision_vals, color='blue', lw=2, label=f'Precision-Recall curve (ELM - {n_neurons} {activation_function})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision-Recall Curve (ELM - {n_neurons} {activation_function})')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.legend(loc="lower left")
    plt.show()

# --- Example Usage of the function with different parameters ---
print("\n--- Running ELM with various hpelm-supported activation functions ---")

# Configuration 1: Original parameters
train_and_evaluate_elm(n_neurons=100, activation_function="sigm",
                       X_train_data=X_train_np, y_train_data=y_train_np,
                       X_test_data=X_test_np, y_test_true=y_test)

# Configuration 2: More neurons, Sigmoid
train_and_evaluate_elm(n_neurons=200, activation_function="sigm",
                       X_train_data=X_train_np, y_train_data=y_train_np,
                       X_test_data=X_test_np, y_test_true=y_test)
# Configuration 3: More neurons, Sigmoid
train_and_evaluate_elm(n_neurons=300, activation_function="sigm",
                       X_train_data=X_train_np, y_train_data=y_train_np,
                       X_test_data=X_test_np, y_test_true=y_test)

# Configuration 4: More neurons, Sigmoid
train_and_evaluate_elm(n_neurons=400, activation_function="sigm",
                       X_train_data=X_train_np, y_train_data=y_train_np,
                       X_test_data=X_test_np, y_test_true=y_test)

# Configuration 5: More neurons, Sigmoid
train_and_evaluate_elm(n_neurons=500, activation_function="sigm",
                       X_train_data=X_train_np, y_train_data=y_train_np,
                       X_test_data=X_test_np, y_test_true=y_test)

# Configuration 6: More neurons, Sigmoid
train_and_evaluate_elm(n_neurons=600, activation_function="sigm",
                       X_train_data=X_train_np, y_train_data=y_train_np,
                       X_test_data=X_test_np, y_test_true=y_test)
```

```python
# --- IMPORTANT: RUN THIS CELL FIRST IN YOUR GOOGLE COLAB NOTEBOOK ---
# This installs all necessary libraries. After it finishes, if prompted,
# RESTART YOUR RUNTIME (Runtime -> Restart runtime) and then run all cells from the beginning.
!pip install joblib imbalanced-learn hpelm scikit-learn matplotlib seaborn pandas numpy

# --- Next, ensure your files are uploaded to your Colab session ---
# Click the folder icon on the left sidebar, then the 'Upload to session storage' icon (paper with an arrow pointing up).
# Make sure you upload:
# 1. Dataset (1).csv (your original training data)
# 2. model_hsbc_hackathon.pkl (your trained model file)
# 3. test.csv (your new test data for prediction, WITHOUT the 'fraud' column)

# --- Main Code for Prediction on New Test Data ---
import pandas as pd
import numpy as np
import joblib # For loading the .pkl model and potentially saving/loading preprocessors
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split # Used for initial data split in setup
from imblearn.over_sampling import SMOTE # Used in initial pipeline setup (if applicable)

# Try importing ELM (needed if your model_hsbc_hackathon.pkl is an ELM)
try:
    from hpelm import ELM
except ImportError:
    print("Warning: 'hpelm' library not found. If your model_hsbc_hackathon.pkl is an ELM, it might fail to load.")
    # Define a dummy class to prevent immediate crash if ELM is not installed and model is ELM
    class DummyELMForLoading:
        def predict(self, X): return np.random.rand(X.shape[0], 1)
    ELM = DummyELMForLoading # Temporarily alias ELM to dummy for loading attempt


print("--- Starting Prediction Process ---")

# --- 1. Load Your Trained Model ---
model_filename = 'model_train_hsbc.pkl'
try:
    # Load the trained model from the .pkl file
    trained_model = joblib.load(model_filename)
    print(f"Successfully loaded trained model from '{model_filename}'")
except FileNotFoundError:
    print(f"Error: Trained model file '{model_filename}' not found. Please upload it to your Google Colab session.")
    exit()
except Exception as e:
    print(f"Error loading trained model from '{model_filename}': {e}")
    print("Please ensure the .pkl file is valid and compatible with your current library versions.")
    exit()


# --- 2. Fit Preprocessing Transformers (consistent with training) ---
# We need to re-fit LabelEncoders and StandardScaler using the original training data
# to ensure consistent transformation of the new test data.
# If you saved your fitted_le_gender.pkl, fitted_le_category.pkl, fitted_scaler.pkl,
# you could load them here instead of refitting.

print("\n--- Fitting Preprocessing Transformers from Original Training Data ---")
try:
    df_train_initial = pd.read_csv('Dataset.csv')
except FileNotFoundError:
    print("Error: 'Dataset.csv' (original training data) not found. Please upload it.")
    exit()

# Define columns used in training (must match your training script)
initial_base_cols = ['step', 'age', 'gender', 'category', 'amount', 'fraud']
df_train_initial = df_train_initial[initial_base_cols].copy()

# Fit LabelEncoders
fitted_le_gender = LabelEncoder()
df_train_initial['gender'] = fitted_le_gender.fit_transform(df_train_initial['gender'])

fitted_le_category = LabelEncoder()
df_train_initial['category'] = fitted_le_category.fit_transform(df_train_initial['category'])

# Feature Engineering (must match training exactly)
if 'amount' in df_train_initial.columns and 'step' in df_train_initial.columns:
    df_train_initial['amount_per_step'] = df_train_initial['amount'] / (df_train_initial['step'] + 1e-6)
    if (df_train_initial['amount'] > 0).all():
        df_train_initial['amount_log'] = np.log1p(df_train_initial['amount'])
    else:
        df_train_initial['amount_log'] = np.log1p(df_train_initial['amount'].clip(lower=0))
elif 'amount' in df_train_initial.columns: # Case if 'step' was missing but 'amount' present
    df_train_initial['amount_per_step'] = 0.0 # Or some other default
    df_train_initial['amount_log'] = np.log1p(df_train_initial['amount'].clip(lower=0))


# Identify numerical columns for StandardScaler (excluding those you explicitly didn't scale)
# This list needs to be consistent with how you scaled in your training script.
# Based on our discussions, often 'age', 'gender', 'category', 'fraud' were excluded.
exclude_from_scaling_train = ['age', 'gender', 'category', 'fraud']
numerical_cols_for_scaling_train = df_train_initial.select_dtypes(include=np.number).columns.drop(exclude_from_scaling_train, errors='ignore').tolist()

fitted_scaler = StandardScaler()
if len(numerical_cols_for_scaling_train) > 0:
    fitted_scaler.fit(df_train_initial[numerical_cols_for_scaling_train])
    print(f"Fitted StandardScaler on columns: {numerical_cols_for_scaling_train}")
else:
    print("No numerical columns were identified for scaling in training (excluding specified).")

print("--- Preprocessing Transformers Fitted ---")


# --- 3. Function to Preprocess New Test Data ---
def preprocess_new_test_data(raw_new_test_df, le_gender_obj, le_category_obj, scaler_obj, scaled_cols_from_training):
    """
    Applies the same preprocessing steps as the training data to new test data.

    Args:
        raw_new_test_df (pd.DataFrame): The raw new test data (without 'fraud' column).
        le_gender_obj (LabelEncoder): Fitted LabelEncoder for 'gender'.
        le_category_obj (LabelEncoder): Fitted LabelEncoder for 'category'.
        scaler_obj (StandardScaler): Fitted StandardScaler object.
        scaled_cols_from_training (list): List of numerical columns that were scaled during training.

    Returns:
        np.ndarray: Preprocessed test data as a NumPy array, ready for prediction.
    """
    processed_df = raw_new_test_df.copy()

    # Ensure raw categorical columns are string type before transformation
    for col in ['gender', 'category']:
        if col in processed_df.columns:
            processed_df[col] = processed_df[col].astype(str)

    # Apply Label Encoding using fitted encoders
    if 'gender' in processed_df.columns:
        # Handle unseen labels in test data by mapping to -1 or a default, or raising error.
        # Here, we'll try to transform and catch errors.
        try:
            processed_df['gender'] = le_gender_obj.transform(processed_df['gender'])
        except ValueError as ve:
            print(f"Warning: Unseen label(s) in 'gender' column during transform: {ve}. Consider handling.")
            # Fallback: map unknown labels to a placeholder like -1 or the mode, or raise.
            # For hackathon, often, just let it error or map to -1 or mode.
            raise # Re-raise to stop if strict handling is needed

    if 'category' in processed_df.columns:
        try:
            processed_df['category'] = le_category_obj.transform(processed_df['category'])
        except ValueError as ve:
            print(f"Warning: Unseen label(s) in 'category' column during transform: {ve}. Consider handling.")
            raise

    # Re-create feature engineering (must be consistent with training)
    if 'amount' in processed_df.columns and 'step' in processed_df.columns:
        processed_df['amount_per_step'] = processed_df['amount'] / (processed_df['step'] + 1e-6)
        if (processed_df['amount'] > 0).all():
            processed_df['amount_log'] = np.log1p(processed_df['amount'])
        else:
            processed_df['amount_log'] = np.log1p(processed_df['amount'].clip(lower=0))
    elif 'amount' in processed_df.columns: # If 'step' is missing but 'amount' present
        processed_df['amount_per_step'] = 0.0 # Or median/mean from training data
        processed_df['amount_log'] = np.log1p(processed_df['amount'].clip(lower=0))
    else: # If amount or step are missing
        processed_df['amount_per_step'] = 0.0
        processed_df['amount_log'] = 0.0 # Or appropriate default

    # Define the final set of features and their order as used by your trained model
    # This list must precisely match the columns of your X_train after preprocessing.
    final_feature_order = ['step', 'age', 'gender', 'category', 'amount', 'amount_per_step', 'amount_log']

    # Ensure all final_feature_order columns are present, fill if missing
    for col in final_feature_order:
        if col not in processed_df.columns:
            processed_df[col] = 0.0 # Fill with default value if missing in test data
            print(f"Warning: Column '{col}' was missing in new test data and filled with 0.")
        else:
            processed_df[col] = pd.to_numeric(processed_df[col], errors='coerce')
            if processed_df[col].isnull().any():
                processed_df[col] = processed_df[col].fillna(0.0) # Fill NaNs created by coercion

    processed_df = processed_df[final_feature_order] # Select and reorder columns

    # Apply StandardScaler using the fitted scaler
    if len(scaled_cols_from_training) > 0:
        cols_to_transform_now = [col for col in scaled_cols_from_training if col in processed_df.columns]
        if len(cols_to_transform_now) > 0:
            processed_df[cols_to_transform_now] = scaler_obj.transform(processed_df[cols_to_transform_now])

    return processed_df.to_numpy().astype(float)


# --- 4. Main Workflow for Prediction on New Test Data ---

new_test_data_filename = 'test.csv' # <<<--- REPLACE WITH YOUR ACTUAL NEW TEST DATA FILE NAME

print(f"\n--- Processing New Test Data from '{new_test_data_filename}' ---")

try:
    raw_new_test_df = pd.read_csv(new_test_data_filename)
    print(f"Loaded new test data. Shape: {raw_new_test_df.shape}")
except FileNotFoundError:
    print(f"Error: New test data file '{new_test_data_filename}' not found. Please upload it to your Colab session.")
    exit()

try:
    # Use the list of columns that were actually scaled during training setup
    preprocessed_new_test_np = preprocess_new_test_data(
        raw_new_test_df,
        fitted_le_gender,
        fitted_le_category,
        fitted_scaler,
        numerical_cols_for_scaling_train # Use the list generated during setup
    )
    print(f"Preprocessed new test data. Shape: {preprocessed_new_test_np.shape}")
except Exception as e:
    print(f"Error during preprocessing new test data: {e}")
    print("Please ensure your test data columns match expected types and values seen in training.")
    exit()

print("Making predictions on new test data using the trained model...")
# Assuming your trained model (e.g., ELM) outputs a "normality score" where lower is anomalous.
# We then do 1 - score to get anomaly_score where higher is anomalous (fraud).
# If your model is a standard classifier (e.g., LightGBM), you might just use .predict_proba()[:, 1]
# For ELM:
if 'hpelm' in sys.modules and isinstance(trained_model, ELM): # Check if hpelm was successfully imported and model is an ELM
    new_test_anomaly_scores = 1 - trained_model.predict(preprocessed_new_test_np).flatten()
    print("Model identified as ELM, predicting anomaly scores.")
else:
    # Fallback for other models or if ELM import failed
    # This assumes your model has a .predict_proba method for classification
    try:
        new_test_anomaly_scores = trained_model.predict_proba(preprocessed_new_test_np)[:, 1]
        print("Model identified as classifier, predicting probabilities.")
    except AttributeError:
        print("Warning: Model does not have .predict_proba. Using .predict() and assuming binary output.")
        new_test_anomaly_scores = trained_model.predict(preprocessed_new_test_np) # Will be 0s and 1s
        # If it's a binary output (0 or 1), ROC AUC might not be meaningful.
        # For anomaly detection, ensure it's a score. If it's 0/1, AUC might be 0.5 or 1.0.


# Apply the best threshold found during your training phase.
# Replace best_threshold_from_training with the actual optimal threshold you determined.
best_threshold_from_training = 0.5 # <<<--- REPLACE WITH YOUR ACTUAL OPTIMAL THRESHOLD (e.g., from ELM tuning)
new_test_predictions_binary = (new_test_anomaly_scores >= best_threshold_from_training).astype(int)

print("\n--- Predictions on New Test Data ---")
print("First 10 Scores (higher value = more likely fraud/anomalous):")
print(new_test_anomaly_scores[:10])
print("\nFirst 10 Binary Predictions (1=Fraud, 0=Not Fraud):")
print(new_test_predictions_binary[:10])

# You can save these predictions to a CSV file
predictions_output_df = pd.DataFrame({
    'original_row_index': raw_new_test_df.index,
    'prediction_score': new_test_anomaly_scores, # Renamed to be generic
    'predicted_fraud': new_test_predictions_binary
})
output_filename = 'new_test_predictions.csv'
predictions_output_df.to_csv(output_filename, index=False)
print(f"\nPredictions saved to '{output_filename}'")

print("\n--- Process Complete ---")
print("Remember to replace the 'trained_model' placeholder with your actual loaded model object.")
print("Also, ensure 'best_threshold_from_training' is set to the optimal value from your training phase.")
print("The 'new_test_data_filename' should point to your test.csv file.")
```
